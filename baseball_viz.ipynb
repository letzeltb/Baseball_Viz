<h1><b>Cleaning Data</h1></b>
Cleaning data

```python
batting.isnull().sum()
```

```python
batting['lgID'] = batting['lgID'].fillna('NULL')

batting['RBI'] = batting['RBI'].fillna(0)
batting['SB'] = batting['SB'].fillna(0)
batting['CS'] = batting['CS'].fillna(0)
batting['SO'] = batting['SO'].fillna(0)
batting['IBB'] = batting['IBB'].fillna(0)
batting['HBP'] = batting['HBP'].fillna(0)
batting['SH'] = batting['SH'].fillna(0)
batting['SF'] = batting['SF'].fillna(0)
batting['GIDP'] = batting['GIDP'].fillna(0)
```

```python
batting['yearID'] = batting['yearID'].astype(str)
batting['teamID'] = batting['teamID'].astype(str)

batting['RBI'] = batting['RBI'].astype(int)
batting['SB'] = batting['SB'].astype(int)
batting['CS'] = batting['CS'].astype(int)
batting['SO'] = batting['SO'].astype(int)
batting['IBB'] = batting['IBB'].astype(int)
batting['HBP'] = batting['HBP'].astype(int)
batting['SH'] = batting['SH'].astype(int)
batting['SF'] = batting['SF'].astype(int)
batting['GIDP'] = batting['GIDP'].astype(int)
```

```python
batting.head(100)
```

<h1><b>Aggregation</h1></b>

```python
yearagg = batting.groupby('yearID',as_index = False).aggregate({'teamID':'nunique','playerID':'nunique','G':'max','AB':'sum','R':'sum','H':'sum','2B':'sum','3B':'sum','BB':'sum','RBI':'sum','HR':'sum','SB':'sum','SO':'sum'})
```

```python
yearagg.tail(20)
```

<h1><b>NORMALIZE</h1></b>

```python
cols_to_norm = ['AB','R','H','2B','3B','BB','RBI','HR','SB','SO']

for col in cols_to_norm:
    yearagg[col] = round(yearagg[col]/yearagg['teamID'],0)
```

```python
yearagg.tail(30)
```

```python
plt.figure(figsize=(18, 10))
sns.heatmap(yearagg.drop(['teamID','playerID'],axis=1).corr(), annot = True, vmin = -1, vmax = 1)
plt.show()
```
